```python
def unique_col_val(df1):
    for col in df1:
        if df1[col].dtypes=='object':
            print(f'{col} : {df1[col].unique()}')
```

```python
unique_col_val(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
unique_col_val(df1)
```

```python
yes_and_no=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
           'DeviceProtection','TechSupport','StreamingTV','StreamingTV','StreamingMovies','PaperlessBilling',
           'Churn']

for col in yes_and_no:
    df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
df1.head()
```

# Gender : Male : 0 --> Female : 1
# Yes --> 1 No--> 0

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.head()
```

```python
df2 = pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.head()
```

```python
df2.dtypes

```

```python
cols = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols]=scaler.fit_transform(df2[cols])
```

```python
df2.sample(5)
```

```python
x=df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test =train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
x_test.shape
```

```python
import tensorflow as tf

from tensorflow import keras
```

```python
## pip install tensorflow == 1.5.0
```

```python
model = keras.Sequential(
[
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid'),
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred =model.predict(x_test)
y_pred[:5]
```

```python
y_pred_new =[]

for i in y_pred:
    if i >0.5:
        y_pred_new.append(1)
    else:
        y_pred_new.append(0)
```

```python
y_pred_new[:5]
```

```python
y_test[:10]
```

```python
y_pred_new[:10]
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_new)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred_new
                           ))
```
